Economic growth of the country is most essential when it comes to the overall development.Economic growth of any country is measured by analysing annual GDP of that specific country with respective time. Estimation of GDP i.e Gross Domestic Product and increse in population, plays vital role while measuring economic growth of country.Let us filter out which economies in asia grew the most.

```python
#Subsetting data of asia from the whole dataset.
AE = gapminder[gapminder['continent'] == "Asia"]
```

```python
AE.head()
```

```python
AE["GDP"]=AE["gdpPercap"] * AE["pop"] 
```

```python
AE
```

```python
df1952= AE[AE['year']== 1952]
df2007= AE[AE['year'] == 2007]
```

```python
#Subset the data according to Years
gapminder1952=df1952[['country','lifeExp','GDP']]
gapminder2007=df2007[['country','lifeExp','GDP']]
```

```python
gapminder1952.sample(10)
```

```python
EInfo = pd.merge(gapminder1952,gapminder2007,how='left' ,on='country')
```

```python
EInfo.sample(10)
```

```python
EInfo=EInfo.rename(columns={"lifeExp_x": "lifeExp52",
                          "lifeExp_y": "lifeExp07","GDP_x":"GDP52","GDP_y":"GDP07"}) 
```

```python
EInfo.sample(10)
```

```python
## Created new column  
EInfo['increase_lfExp'] = EInfo['lifeExp07'] - EInfo['lifeExp52']
EInfo['increase_GDP'] = EInfo['GDP07'] - EInfo['GDP52']
```

```python
#subseting the data according to the GDP and Life expactancy increase.
GrowthWiseLifeExp=EInfo[['country','increase_lfExp']]
GrowthWiseGDP=EInfo[['country','increase_GDP']]

#data sorted accordingly.
GrowthWiseLifeExpSorted = GrowthWiseLifeExp.sort_values('increase_lfExp', ascending=False)
GrowthWiseGDPSorted = GrowthWiseGDP.sort_values('increase_GDP', ascending=False)

#top 5 countries taken out for the dataset.
GrowthWiseLifeExp1st5=GrowthWiseLifeExpSorted.iloc[0:5]
GrowthWiseGDP1st5=GrowthWiseGDPSorted.iloc[0:5]

#index reset.
GrowthWiseLifeExp1st5=GrowthWiseLifeExp1st5.reset_index()
GrowthWiseGDP1st5=GrowthWiseGDP1st5.reset_index()


GrowthWiseLifeExp1st5= GrowthWiseLifeExp1st5.drop(['index'], axis=1)
GrowthWiseGDP1st5= GrowthWiseGDP1st5.drop(['index'], axis=1)
```

#Which economies in Asia grew the most in relative terms ?
Defining growth in terms of incremental evolution of GDP and life expectancy

```python
UniYear = pd.DataFrame(index=AE.year.unique())

for count in (AE.country.unique()):
    df = AE[AE['country']==count]
    df = (df.drop(['country','continent'],axis = 1).sort_values(by='year'))
    df[count] = df['pop']*df['gdpPercap']
    
    df.set_index('year',inplace = True)
    init_Val = df[count].iloc[0]
    df_GDP =  df[count]/init_Val * 100

    UniYear = UniYear.join(df_GDP)
```

```python
list_top_10 = list(((UniYear.T[2007].sort_values().tail(10)).to_frame()).index)
UniYear[list_top_10].plot(kind='line', subplots=False,figsize=(10,10))
```

As illustrated in above graph economy of Singapore is showing maximum Economic growth in relative terms followed by economy of Oman.
Is there any relationship between GDP Per Capita and life Expectancy ?

```python
sns.regplot(x="gdpPercap", y="lifeExp", data=gapminder, color = 'red');
```

Yes, there is relationship between GDP per Capita and Life Expectancy as GDP increases Life Expectancy also increase. We can also see there that there almost none of the country with gdpPercap more than the 20000 has life expactancy less than 40 years.

```python
sns.jointplot(
    data=gapminder,
    x="gdpPercap", y="lifeExp", hue="year",
    kind="kde",
)
```

We can see the same interpretation using the joint kernel density estimate visualization

```python
mlp = sns.FacetGrid(gapminder,row="year")

mlp.map(sns.scatterplot, "lifeExp","gdpPercap")
```

Yes, relationship between GDP Per Capita and Life Expectancy has been changed over time. Observing this relationship from 1952 its clear that rise in GDP and Life Expectancy showing regular growth.
